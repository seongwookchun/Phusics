```python
import requests
import time
from random import randint
from time import sleep
from IPython.core.display import clear_output
start_time = time.time()
#timestart_time = time()
rquests = 0
tot_pages = 12
pages = [str(i) for i in range(1,tot_pages +1)]
soups = []

df_final = pd.DataFrame()
for i in range(tot_pages):
    # A request would go here
    rquests += 1
    sleep(randint(1,2))
    elapsed_time = time.time() - start_time
    print('Request: {}; Frequency: {} requests/s'.format(rquests, round(rquests/elapsed_time, 4)))
    
    url = "http://www.worlditshow.co.kr/kor/search/product_list.asp?page=" + str(pages[i])
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    rowContainer = soup.find_all('tr')
    
    for i in range(1, len(rowContainer)):
        temp01 = rowContainer[i].td.text
        temp02 = rowContainer[i].td.find_next('td').text
        temp03 = rowContainer[i].td.find_next('td').find_next('td').text
        temp04 = rowContainer[i].td.find_next('td').find_next('td').find_next('td').text
        temp05 = rowContainer[i].td.find_next('td').find_next('td').find_next('td').find_next('td').text

        df_final = df_final.append({ '번호': temp01 , '부스번호' : temp02, '회사명' : temp03, '홈페이지': temp04,
                                     '주 전시품목': temp05}, ignore_index=True)
    print('a spoon of soup has been scooped')
    clear_output(wait = True)
    
print('soup has been cooked')
display(df_final)
```
